## Compare accuracies: SSL vs SL approaches

```python
import matplotlib
import matplotlib.pyplot as plt
# plt.rcParams["figure.figsize"] = (9, 8)
# matplotlib.style.use("default")

SAMPLES_PER_CLASS = [10, 20, 30, 40, 50, 80, 110, 140, 170, 200]
layers = ["add_3", "add_2", "conv2d_6"]

# self-supervised
add_19_accs_ssl = [68.636, 77.349, 81.225, 82.946, 85.349, 88.233, 90.124, 92.062, 91.628, 92.295]
add_18_accs_ssl = [66.713, 78.434, 83.256, 85.783, 86.915, 90.248, 92.806, 93.535, 94.481, 95.302]
conv2d_35_accs_ssl = [52.744, 59.519, 62.729, 65.752, 68.078, 69.659, 70.915, 70.977, 72.605, 73.736]

# supervised
add_19_accs_sl = [71.333, 82.543, 85.349, 88.388, 90.527, 92.419, 94.341, 95.566, 96.264, 96.868]
add_18_accs_sl = [70.341, 80.45, 86.0, 87.984, 89.566, 93.845, 94.636, 95.984, 96.0, 96.915]
conv2d_35_accs_sl = [64.248, 73.969, 78.109, 79.876, 82.651, 85.674, 86.682, 87.984, 88.806, 89.395]

# for i in range(len(layers)):
plt.figure(figsize=(9, 8))
plt.ylim(50, 100)
plt.plot(SAMPLES_PER_CLASS, add_19_accs_ssl, marker='o', markersize=10, linestyle='--', color='red', label=layers[0] + ": SSL")
plt.plot(SAMPLES_PER_CLASS, add_18_accs_ssl, marker='o', markersize=10, linestyle='--', color='maroon', label=layers[1] + ": SSL")
plt.plot(SAMPLES_PER_CLASS, conv2d_35_accs_ssl, markersize=10, marker='o', linestyle='--', color='orange', label=layers[2] + ": SSL")

plt.plot(SAMPLES_PER_CLASS, add_19_accs_sl, marker='x', markersize=10, linestyle='solid', color='forestgreen', alpha=1, label=layers[0] + ": SL")
plt.plot(SAMPLES_PER_CLASS, add_18_accs_sl, marker='x', markersize=10, linestyle='solid', color='lime', alpha=1, label=layers[1] + ": SL")
plt.plot(SAMPLES_PER_CLASS, conv2d_35_accs_sl, marker='x', markersize=10, linestyle='solid', color='mediumspringgreen', alpha=1, label=layers[2] + ": SL")

plt.legend(fontsize=16, ncol=2)
plt.xlabel("Samples per Class", fontsize=22)
plt.ylabel("Test Accuracy", fontsize=22) 
plt.grid(alpha=0.2)
plt.xticks(SAMPLES_PER_CLASS, fontsize=18, rotation=45)
plt.yticks(fontsize=18)
plt.savefig('minixception_accs_ssl_vs_sl.pdf', bbox_inches='tight')
    
plt.show()
```
