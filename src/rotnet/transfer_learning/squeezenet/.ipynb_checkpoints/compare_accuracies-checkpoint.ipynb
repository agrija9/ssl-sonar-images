## Compare accuracies: SSL vs SL approaches

```python
import matplotlib
import matplotlib.pyplot as plt
# plt.rcParams["figure.figsize"] = (9, 8)
# matplotlib.style.use("default")

SAMPLES_PER_CLASS = [10, 20, 30, 40, 50, 80, 110, 140, 170, 200]
# layers = ["batch_norm_8", "batch_norm_9", "global_avg_pooling2d"]
layers = ["batch_norm_8", "batch_norm_9", "global_avg_pool"]

# self-supervised
batch_norm_8_accs_ssl = [62.543, 72.093, 77.69, 82.109, 82.388, 87.023, 89.566, 90.481, 91.674, 92.372]
batch_norm_9_accs_ssl = [39.705, 47.984, 52.744, 55.147, 56.946, 59.302, 61.643, 62.682, 62.264, 64.651]
global_average_pooling2d_accs_ssl = [23.922, 24.434, 25.38, 25.132, 25.969, 25.519, 26.279, 25.907, 25.845, 25.953]

# supervised
batch_norm_8_accs_sl = [71.767, 79.054, 85.705, 87.442, 89.659, 92.295, 93.674, 95.147, 95.938, 96.31]
batch_norm_9_accs_sl = [56.078, 66.326, 71.38, 73.597, 76.202, 78.698, 80.341, 81.752, 82.0, 83.07]
global_average_pooling2d_accs_sl = [38.775, 42.372, 43.302, 45.349, 45.643, 47.488, 47.225, 47.86, 48.372, 48.202]

# for i in range(len(layers)):
plt.figure(figsize=(9, 8))
plt.ylim(20, 100)
plt.plot(SAMPLES_PER_CLASS, batch_norm_8_accs_ssl, marker='o', markersize=10, linestyle='--', color='red', label=layers[0] + ": SSL")
plt.plot(SAMPLES_PER_CLASS, batch_norm_9_accs_ssl, marker='o', markersize=10, linestyle='--', color='maroon', label=layers[1] + ": SSL")
plt.plot(SAMPLES_PER_CLASS, global_average_pooling2d_accs_ssl, markersize=10, marker='o', linestyle='--', color='orange', label=layers[2] + ": SSL")

plt.plot(SAMPLES_PER_CLASS, batch_norm_8_accs_sl, marker='x', markersize=10, linestyle='solid', color='forestgreen', alpha=1, label=layers[0] + ": SL")
plt.plot(SAMPLES_PER_CLASS, batch_norm_9_accs_sl, marker='x', markersize=10, linestyle='solid', color='lime', alpha=1, label=layers[1] + ": SL")
plt.plot(SAMPLES_PER_CLASS, global_average_pooling2d_accs_sl, marker='x', markersize=10, linestyle='solid', color='mediumspringgreen', alpha=1, label=layers[2] + ": SL")

plt.legend(fontsize=16, ncol=2)
plt.xlabel("Samples per Class", fontsize=22)
plt.ylabel("Test Accuracy", fontsize=22) 
plt.grid(alpha=0.2)
plt.xticks(SAMPLES_PER_CLASS, fontsize=18, rotation=45)
plt.yticks(fontsize=18)
plt.savefig('squeezenet_accs_ssl_vs_sl.pdf', bbox_inches='tight')

plt.show()
```
