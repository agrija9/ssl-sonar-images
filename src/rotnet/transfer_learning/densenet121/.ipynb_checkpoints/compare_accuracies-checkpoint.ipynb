## Compare accuracies: SSL vs SL approaches

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
# plt.rcParams["figure.figsize"] = (9, 8)
# matplotlib.style.use("default")

SAMPLES_PER_CLASS = [10, 20, 30, 40, 50, 80, 110, 140, 170, 200]
# layers = ["conv5_block15_0_relu", "conv5_block16_0_relu", "avg_pool"]
layers = ["conv5_block15", "conv5_block16", "avg_pool"]

# self-supervised
conv5_block15_accs_ssl = [67.953, 77.395, 84.279, 85.597, 87.566, 90.853, 92.109, 93.953, 94.465, 95.023]
e = np.array([2.174, 1.978, 0.956, 1.202, 0.906, 0.8, 0.905, 0.58, 0.405, 0.435])

conv5_block16_accs_ssl = [68.853, 78.977, 83.829, 86.899, 88.186, 91.039, 92.589, 93.907, 94.791, 94.667]
avg_pool_accs_ssl  = [64.512, 72.667, 78.279, 80.961, 83.736, 86.31, 88.264, 88.868, 90.357, 90.171]

# supervised
conv5_block15_accs_sl = [69.953, 79.054, 83.008, 86.698, 88.729, 91.783, 93.752, 94.543, 95.008, 95.705]
conv5_block16_accs_sl = [68.977, 79.55, 83.969, 86.264, 89.287, 91.783, 93.566, 94.372, 95.318, 95.612]
avg_pool_accs_sl = [63.659, 73.721, 78.76, 81.829, 83.287, 87.101, 88.605, 90.047, 90.868, 91.969]

# for i in range(len(layers)):
plt.figure(figsize=(9, 8))
plt.ylim(60, 100)
plt.plot(SAMPLES_PER_CLASS, conv5_block15_accs_ssl, marker='o', markersize=10, linestyle='--', color='red', label=layers[0] + ": SSL")
plt.plot(SAMPLES_PER_CLASS, conv5_block16_accs_ssl, marker='o', markersize=10, linestyle='--', color='maroon', label=layers[1] + ": SSL")
plt.plot(SAMPLES_PER_CLASS, avg_pool_accs_ssl, markersize=10, marker='o', linestyle='--', color='orange', label=layers[2] + ": SSL")

plt.plot(SAMPLES_PER_CLASS, conv5_block15_accs_sl, marker='x', markersize=10, linestyle='solid', color='forestgreen', alpha=1, label=layers[0] + ": SL")
plt.plot(SAMPLES_PER_CLASS, conv5_block16_accs_sl, marker='x', markersize=10, linestyle='solid', color='lime', alpha=1, label=layers[1] + ": SL")
plt.plot(SAMPLES_PER_CLASS, avg_pool_accs_sl, marker='x', markersize=10, linestyle='solid', color='mediumspringgreen', alpha=1, label=layers[2] + ": SL")

plt.legend(fontsize=16, ncol=2)
plt.xlabel("Samples per Class", fontsize=22)
plt.ylabel("Test Accuracy", fontsize=22) 
plt.grid(alpha=0.2)
plt.xticks(SAMPLES_PER_CLASS, fontsize=18, rotation=45)
plt.yticks(fontsize=18)
plt.savefig('densenet121_accs_ssl_vs_sl.pdf', bbox_inches='tight')

plt.show()
```
