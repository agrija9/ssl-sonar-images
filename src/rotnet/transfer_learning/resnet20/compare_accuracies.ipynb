## Compare accuracies: SSL vs SL approaches

```python
import matplotlib
import matplotlib.pyplot as plt
# plt.rcParams["figure.figsize"] = (9, 8)
# matplotlib.style.use("default")

SAMPLES_PER_CLASS = [10, 20, 30, 40, 50, 80, 110, 140, 170, 200]
layers = ['flatten', 'activation_18', 'activation_17']

# self-supervised
flatten_accs_ssl = [68.217, 77.922, 81.674, 84.403, 85.643, 89.318, 90.062, 91.659, 92.093, 92.078]
activation_18_accs_ssl = [66.636, 77.054, 84.031, 85.705, 87.008, 91.566, 93.488, 95.287, 95.814, 96.465]
activation_17_accs_ssl = [66.186, 77.876, 82.124, 85.287, 87.891, 90.698, 93.628, 95.209, 95.457, 96.62]

# supervised
flatten_accs_sl = [76.512, 84.868, 87.395, 90.202, 90.698, 93.054, 94.62, 95.597, 95.767, 96.171]
activation_18_accs_sl = [70.14, 79.814, 85.628 , 88.062 , 90.31, 93.008, 95.271, 95.69, 96.713 , 97.271]
activation_17_accs_sl = [71.612, 81.24, 86.326, 88.434, 89.752, 92.636, 95.256, 95.814, 96.357, 97.147]

# for i in range(len(layers)):
plt.figure(figsize=(9, 8))
plt.ylim(60, 100)
plt.plot(SAMPLES_PER_CLASS, flatten_accs_ssl, marker='o', markersize=10, linestyle='--', color='red', label=layers[0] + ": SSL")
plt.plot(SAMPLES_PER_CLASS, activation_18_accs_ssl, marker='o', markersize=10, linestyle='--', color='maroon', label=layers[1] + ": SSL")
plt.plot(SAMPLES_PER_CLASS, activation_17_accs_ssl, markersize=10, marker='o', linestyle='--', color='orange', label=layers[2] + ": SSL")

plt.plot(SAMPLES_PER_CLASS, flatten_accs_sl, marker='x', markersize=10, linestyle='solid', color='forestgreen', alpha=1, label=layers[0] + ": SL")
plt.plot(SAMPLES_PER_CLASS, activation_18_accs_sl, marker='x', markersize=10, linestyle='solid', color='lime', alpha=1, label=layers[1] + ": SL")
plt.plot(SAMPLES_PER_CLASS, activation_17_accs_sl, marker='x', markersize=10, linestyle='solid', color='mediumspringgreen', alpha=1, label=layers[2] + ": SL")


plt.legend(fontsize=16, ncol=2)
plt.xlabel("Samples per Class", fontsize=22)
plt.ylabel("Test Accuracy", fontsize=22) 
plt.grid(alpha=0.2)
plt.xticks(SAMPLES_PER_CLASS, fontsize=18, rotation=45)
plt.yticks(fontsize=18)
plt.savefig('resnet20_accs_ssl_vs_sl.pdf', bbox_inches='tight')

plt.show()
```
