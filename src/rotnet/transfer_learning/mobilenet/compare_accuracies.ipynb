## Compare accuracies: SSL vs SL approaches

```python
import matplotlib
import matplotlib.pyplot as plt
# plt.rcParams["figure.figsize"] = (9, 8)
# matplotlib.style.use("default")

SAMPLES_PER_CLASS = [10, 20, 30, 40, 50, 80, 110, 140, 170, 200]
# layers = ["conv_pw_11_relu", "flatten", "conv_pw_12_relu"]
layers = ["conv_pw_11", "flatten", "conv_pw_12"]

# self-supervised
conv_pw_11_relu_accs_ssl = [58.217, 67.147, 71.318, 74.667, 77.194, 81.891, 83.86, 85.039, 86.915, 88.14]
flatten_5_accs_ssl = [51.085, 58.357, 64.186, 68.326, 70.0, 73.426, 76.682, 78.217, 80.357, 82.357]
conv_pw_12_relu_accs_ssl = [54.977, 62.372, 67.333, 70.636, 73.721, 77.039, 79.302, 81.24, 83.101, 83.318]


# supervised
conv_pw_11_relu_accs_sl = [63.907, 70.899, 72.62, 76.388, 78.202, 81.488, 84.093, 85.55, 87.023, 87.349]
flatten_5_accs_sl = [60.047, 66.264, 69.984, 72.109, 74.295, 77.209, 78.806, 79.953, 81.147, 81.643]
conv_pw_12_relu_accs_sl = [59.984, 67.891, 72.45, 74.357, 76.233, 78.729, 80.403, 82.465, 83.395, 83.69]

# for i in range(len(layers)):
plt.figure(figsize=(9, 8))
plt.ylim(50, 100)
plt.plot(SAMPLES_PER_CLASS, conv_pw_11_relu_accs_ssl, marker='o', markersize=10, linestyle='--', color='red', label=layers[0] + ": SSL")
plt.plot(SAMPLES_PER_CLASS, flatten_5_accs_ssl, marker='o', markersize=10, linestyle='--', color='maroon', label=layers[1] + ": SSL")
plt.plot(SAMPLES_PER_CLASS, conv_pw_12_relu_accs_ssl, markersize=10, marker='o', linestyle='--', color='orange', label=layers[2] + ": SSL")

plt.plot(SAMPLES_PER_CLASS, conv_pw_11_relu_accs_sl, marker='x', markersize=10, linestyle='solid', color='forestgreen', alpha=1, label=layers[0] + ": SL")
plt.plot(SAMPLES_PER_CLASS, flatten_5_accs_sl, marker='x', markersize=10, linestyle='solid', color='lime', alpha=1, label=layers[1] + ": SL")
plt.plot(SAMPLES_PER_CLASS, conv_pw_12_relu_accs_sl, marker='x', markersize=10, linestyle='solid', color='mediumspringgreen', alpha=1, label=layers[2] + ": SL")

plt.legend(fontsize=16, ncol=2)
plt.xlabel("Samples per Class", fontsize=22)
plt.ylabel("Test Accuracy", fontsize=22) 
plt.grid(alpha=0.2)
plt.xticks(SAMPLES_PER_CLASS, fontsize=18, rotation=45)
plt.yticks(fontsize=18)
plt.savefig('mobilenet_accs_ssl_vs_sl.pdf', bbox_inches='tight')

plt.show()
```
