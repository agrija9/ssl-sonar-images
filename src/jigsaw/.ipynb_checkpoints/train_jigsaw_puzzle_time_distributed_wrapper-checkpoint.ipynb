## define original (raw) watertank dataset

```python
class SonarDebrisSupervised(object):
    def __init__(self, file_path):
        self.file_path = file_path

    def get_sonar_data(self):

        """
        Reads from HDF5 file containing sonar data (resized to fix dims).
        Returns list of np arrays containing image data.
        """

        print("[INFO] Retrieving Sonar Debris Supervised Data")
        with h5py.File(self.file_path, "r") as f:
            # get images and labels
            x_train = f["x_train"][...].astype(np.float32)
            y_train = f["y_train"][...]

            x_val = f["x_val"][...].astype(np.float32)
            y_val = f["y_val"][...]

            x_test = f["x_test"][...].astype(np.float32)
            y_test = f["y_test"][...]

            print("[INFO] Data dimensions")
            print("Train", len(x_train))
            print("Val", len(x_val))
            print("Test", len(x_test))

        return (x_train, y_train), (x_val, y_val), (x_test, y_test)

def load_sonar_debris_supervised(file_path, batch_size):
    """
    Tensorflow data pipeline: loads data as numpy arrays, defines tf dataset, performs
    image augmentation (random shift and random flipping), returns model-ready data.
    """

    print()
    print("[INFO] Loading Tf datasets")

    dataset_object = SonarDebrisSupervised(file_path)

    # Read data
    (x_train, y_train), (x_val, y_val), (x_test, y_test) = dataset_object.get_sonar_data()
    
    return (x_train, y_train), (x_val, y_val), (x_test, y_test)
```

## define wild sonar dataset

```python
# comes with no labels
class SonarWildData(object):
    def __init__(self, file_path):
        self.file_path = file_path

    def get_sonar_data(self):
        """
        Reads from HDF5 file containing sonar data (resized to fix dims).
        Returns list of np arrays containing image data.
        """
        random.seed(10)

        print("[INFO] Retrieving Wild Sonar Data")
        with h5py.File(self.file_path, "r") as f:
            print("Keys: %s" % f.keys())
            a_group_key = list(f.keys())[0]
            print(f[a_group_key])

            # Get wild data # NOTE: casting to list takes way too long
            data = np.array(f[a_group_key]).astype(np.float32)
            
            # generate train/val/test sets
            # NOTE: downsampling dataset by half (63K/2) due to memory issues
            x_train = data[:int(len(data)*0.7),:]
            x_val = data[int(len(data)*0.7):int(len(data)*0.7)+int(len(data)*0.15),:]
            x_test = data[int(len(data)*0.7)+int(len(data)*0.15):,:]
            
            print("[INFO] Original Data dimensions")
            print("Train", len(x_train))
            print("Val", len(x_val))
            print("Test", len(x_test))
            print()
            
            # 1/2 too big, testing 1/10
            x_train = x_train[:int(len(x_train)/10)]
            x_val = x_val[:int(len(x_val)/10)]
            x_test = x_test[:int(len(x_test)/10)]
            
            print("[INFO] Downsampled Data dimensions")
            print("Train", len(x_train))
            print("Val", len(x_val))
            print("Test", len(x_test))

        return (x_train, _), (x_val, _), (x_test, _)

def load_wild_sonar_data(file_path, batch_size):
    print()
    print("[INFO] Loading Tf datasets")

    dataset_object = SonarWildData(file_path)

    # Read data
    (x_train, _), (x_val, _), (x_test, _) = dataset_object.get_sonar_data()
    
    return (x_train, _), (x_val, _), (x_test, _)
```

## define model

```python
def build_feature_extractor(shape=(32, 32, 1)):
    """
    """
    feature_extractor = keras.models.Sequential()
    
    feature_extractor.add(keras.Input(shape=shape))

    feature_extractor.add(Conv2D(32, kernel_size=(3,3), padding='same', activation='relu'))  # 96
    feature_extractor.add(BatchNormalization())
    feature_extractor.add(MaxPooling2D((2,2), padding='same'))
    feature_extractor.add(Dropout(0.3))
    
    feature_extractor.add(Conv2D(16, kernel_size=(3,3), padding='same', activation='relu')) # 256
    feature_extractor.add(BatchNormalization())
    feature_extractor.add(MaxPooling2D((2,2), padding='same'))
    feature_extractor.add(Dropout(0.3))

    feature_extractor.add(Conv2D(8, kernel_size=(3,3), padding='same', activation='relu')) # 384
    feature_extractor.add(BatchNormalization())
    feature_extractor.add(MaxPooling2D((2,2), padding='same'))
    feature_extractor.add(Dropout(0.3))
    
    feature_extractor.add(Flatten())
    # feature_extractor.add(Dense(128, activation="relu")) 
    
    return feature_extractor
    
def build_time_distributed_model(shape=(9, 32, 32, 1), no_classes=10):
    """
    """
    # define feature extractor
    feature_extractor = build_feature_extractor(shape[1:])
    
    # define time distributed layer
    model = keras.models.Sequential()
    model.add(TimeDistributed(feature_extractor, input_shape=shape))
    
    # decision network
    model.add(Flatten())
    model.add(Dense(64, activation='relu')) # 256, 128, 64
    model.add(BatchNormalization())
    model.add(Dense(32, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.3))
    model.add(Dense(no_classes, activation="softmax"))
              
    return model, feature_extractor
```

## generate puzzles from sonar data (this generates a few GB of data)

* this goes to jigsaw_gen.py
* store data as jpg files (image + labels)

```python
def generate_puzzles(data_array, channels, image_dim, 
                     cuts, out_dir, no_permutations, 
                     target_permutations, dataset_name):
    """
    channels : image channels (1 for grayscale)
    image_dim : dimensions of input image (assuming square image)
    cuts : number of tiles per image (2 or 3 puzzles)
    no_permutations : times a single image is shuffled into a puzzle (we want 10)
    dataset_name : train / valid / test
    """
    ind = -1

    # dictionary to keep track of image-labels into csv file
    # data = {'image': [], 'label': []}
    data = {"image": [], "permutation order": [], "label": []}
    
    # iterate over each permutation order for all images
    for permutation_index in range(no_permutations):
        
        rand_pos = target_permutations[permutation_index]
        
        cut_len = image_dim // cuts

        if cut_len*cuts != image_dim:
            return "provide a valid cut value"
        
        for image in data_array:
            ind += 1

            if image.shape != (image_dim, image_dim, channels):
                continue

            new_img = np.zeros((image_dim, image_dim, channels))
            
            # rand_pos = np.random.permutation([i for i in range(cuts*cuts)])
            
            for i in range(cuts):
                hor_cut = image[i*cut_len:(i+1)*cut_len]

                for j in range(cuts):
                    piece = hor_cut[:, j*cut_len:(j+1)*cut_len]
                    pos = rand_pos[i*cuts+j]
                    x = pos//cuts
                    y = pos%cuts
                    new_img[x*cut_len:(x+1)*cut_len,y*cut_len:(y+1)*cut_len] = piece

            # label = np.zeros(cuts*cuts).astype('int')

            # for i, val in enumerate(rand_pos):
                # label[val] = int(i)

            # new_label = ''
            # for i in label:
                # new_label += str(i) + ' '
            
            permutation_order = np.zeros(cuts*cuts).astype('int')

            for i, val in enumerate(rand_pos):
                permutation_order[val] = int(i)

            log_permutation_order = ''
            for i in permutation_order:
                log_permutation_order += str(i) + ' '
                
            data['image'].append(str(ind)+'.jpg')
            data['permutation order'].append(log_permutation_order)
            data["label"].append(permutation_index)
            
            # np.save(out_dir + str(ind) + '.jpg', new_im)
            # saved_img = Image.fromarray(new_im)
            # saved_img.save(out_dir + str(ind) + ".jpg")
            # new_img = new_img.astype(np.uint8)
            # I8 = (((I - I.min()) / (I.max() - I.min())) * 255.9).astype(np.uint8)
            # img = Image.fromarray(I8)
            # img.save("file.png")
            # plt.imshow(new_img, cmap="gray")
            plt.imsave(out_dir + "/" + str(ind) + '.jpg', np.squeeze(new_img, axis=2), cmap="gray")
    
    # store label data in csv file
    df = pd.DataFrame(data)
    df.to_csv("sonar_puzzle_data/puzzle_{}x{}/{}.csv".format(cuts, cuts, dataset_name))

# if __name__ == '__main__':
```

```python
def rearrange(im, label, cuts=2, dim=96, channel=3):
    """
    rearranges the image according to the given label
    """

    cut_len = dim // cuts # e.g. 32
    new_im = np.zeros((dim, dim, channel))
    
    for i in range(cuts):
        hor_cut = im[i*cut_len : (i + 1)*cut_len] # 0:32, 32:64, 64:96 
        for j in range(cuts):
            piece = hor_cut[:, j*cut_len : (j + 1)*cut_len] # piece (32, 32, 1), indices 0:32, 32:64, 64:96 (3 times)
            # print(j*cut_len, (j + 1) * cut_len)
            # print(piece)    
            pos = label[i * cuts + j] # label[0],...,label[8]
            print(pos)
            # print(i * cuts + j)
            x = pos // cuts
            y = pos % cuts
            # print("x:", x)
            # print("y:", y)
            # print()
            new_im[x * cut_len:(x + 1) * cut_len, y * cut_len:(y + 1) * cut_len] = piece

    # plt.imshow(new_im)
    # plt.show()
    return new_im
    
def extract_piece(array, size=96, cuts=2):
    """
    extracts each piece of the puzzle and returns cut image (4 segments)
    """

    cut_len = size // cuts

    if cuts == 3:
        array = np.array([array[:, 0:cut_len, :], array[:, cut_len:cut_len * 2, :], array[:, cut_len * 2:cut_len * 3, :]])
        array = np.concatenate(
            (array[:, 0:cut_len, :, :], array[:, cut_len:cut_len * 2, :, :], array[:, cut_len * 2:cut_len * 3, :, :]))
    if cuts == 2:
        array = np.array([array[:, 0:cut_len, :], array[:, cut_len:, :]])
        array = np.concatenate((array[:, 0:cut_len, :, :], array[:, cut_len:, :, :]))

    return array

def load_data(base_data_path, data_set, cuts=2):
    """
    loads and returns data
    """
    
    data = pd.read_csv(base_data_path + '{}.csv'.format(data_set))
    x = []
    y = []
    total = len(data)
    
    for i in range(total):
        im = Image.open(base_data_path + data_set + "/" + data.iloc[i]['image'])
        im = np.array(im).astype(np.float32)
        im = im/255
        # im = im / 255 - 0.5
        
        if data_set  == "test":
        # if path.split('/')[-2] == 'test':
            x.append(im)
        else:
            x.append(extract_piece(im, size=96, cuts=cuts))

        label = data.iloc[i]['label']
        # label = [int(i) for i in label.split()]
        y.append(label)

    return (np.array(x), np.expand_dims(np.array(y), axis=-1))
```

## compute accuracies

```python
x_test_puzzle_processed = []

for sample in x_test_puzzle:
    x_test_puzzle_processed.append(extract_piece(sample, size=96, cuts=cuts))
    
x_test_puzzle_processed = np.array(x_test_puzzle_processed)
x_test_puzzle_processed = x_test_puzzle_processed[:,:,:,:,:1]

print("Test data shape: ", x_test_puzzle.shape)
print("Test data model shape: ", x_test_puzzle_processed.shape)

# compute predictions
y_pred = model.predict(x_test_puzzle_processed)

print()
print("Predictions shape: ", y_pred.shape)
print("True labels shape: ", y_test_puzzle_tf.shape)

# compute accuracy percentage
accuracy = np.mean(np.equal(np.argmax(y_pred, axis=-1), np.argmax(y_test_puzzle_tf, axis=-1)))*100
print("Accuracy on {} test samples: {}%".format(len(y_pred), round(accuracy, 2)))

with open("trained_models/{}x{}_puzzle_{}_epochs_{}_perms.txt".format(cuts, cuts, EPOCHS, no_permutations), "a") as text_file:
    print("Predictions shape: ", y_pred.shape, file=text_file)
    print("True labels shape: ", y_test_puzzle_tf.shape, file=text_file)
    print("Accuracy on {} test samples: {}%".format(len(y_pred), round(accuracy, 2)), file=text_file)
```
